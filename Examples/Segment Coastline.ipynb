```python
img_segm = img_segm.astype('uint8')
#plt.imshow(img_segm)
```

### Adaptive Threshold

```python
thresh = cv2.adaptiveThreshold(img_segm,np.max(img_segm)/2,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,17,0)

plt.figure(figsize=(30,30))
plt.gray()

plt.title('Regional Adaptive Segmentation Result', fontsize = 32)
plt.imshow(thresh)
```

```python
small = morphology.remove_small_objects(thresh.astype('bool'), 100000)
plt.figure(figsize=(30,30))
plt.imshow(small)
```

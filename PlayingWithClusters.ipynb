# Let's Cluster

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=5, random_state=0).fit(pic)
pic2show = kmeans.cluster_centers_[kmeans.labels_]
```

```python
kmeans = KMeans(n_clusters=3, random_state=0).fit(pic)
pic2show = kmeans.cluster_centers_[kmeans.labels_]
```

```python
plt.imshow(pic2show)
```

```python
kmeans = KMeans(n_clusters=7, random_state=0).fit(pic)
pic2show = kmeans.cluster_centers_[kmeans.labels_]
```

```python
plt.imshow(pic2show)
```
